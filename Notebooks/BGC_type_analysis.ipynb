```python
from pathlib import Path
from Bio import SeqIO
from collections import defaultdict 
import pandas as pd
import venn
import upsetplot
```

```python
folder = Path("../Data/mibig/")

cluster_type_dict = defaultdict(int)

cluster_type_dict_T1PKS_only = defaultdict(int)

df_dict_list = []
for i, filename in enumerate(folder.glob("*.gbk")):
    gb_list = []
    cluster_annotations = []
    for gb_record in SeqIO.parse(open(filename, "r"), "genbank"):
        gb_list.append(gb_record)
    if len(gb_list)>1:
        raise ValueError
    gb = gb_list[0]
    
    print(filename, len(gb_list))
    is_T1PKS = False
    for f in gb.features:
        if f.type == "region":
            print(f.qualifiers["product"])
            if "T1PKS" in f.qualifiers["product"]:
                is_T1PKS = True

    for f in gb.features:
        try:
            for l in f.qualifiers["NRPS_PKS"]:
                if l[:4]=="type":
                    cluster_annotations.append(l.split(":")[-1].strip())
        except KeyError:
            pass
    dic = {}
    for c in set(cluster_annotations):
        cluster_type_dict[c]+= 1
        dic[c] = True
        if is_T1PKS:
            cluster_type_dict_T1PKS_only[c]+=1

    df_dict_list.append(dic)
    cluster_type_dict["total"]+= 1
    if is_T1PKS:
        cluster_type_dict_T1PKS_only["total"]+=1
```

```python
nrps_pk_df = pd.DataFrame(df_dict_list)
nrps_pk_df = nrps_pk_df.fillna(False)
```
